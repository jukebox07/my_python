```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
my_url="https://www.flipkart.com/furniture/mattresses/pr?sid=wwe%2Crg9&otracker=nmenu_sub_Home+%26+Furniture_0_Mattresses&sort=price_asc"
uClient=uReq(my_url)
page_html=uClient.read()
uClient.close()
```

```python
page_html
```

```python
page_sour=soup(page_html,"html.parser")
containers=page_sour.findAll("div",{"class":"_1AtVbE col-12-12"})
print(len(containers))
```

```python
containers=page_sour.findAll("div",{"class":"_1AtVbE"})
print(len(containers))
print(soup.prettify(containers[0]))
```

```python
price=page_sour.findAll('a',{'class':'_8VNy32'})
print(price[0].text)
```

```python
products=[]
pricing=[]
description=[]

for contain in containers:
    page_sour.findAll('a',{'class':'_2rpwqI'}) 
    product_name=tat[0].img['alt']
    products.append(product_name)
    price_container=price=page_sour.findAll('a',{'class':'_8VNy32'})
    price= price_container[0].text.strip()
    pricing.append(price)
    desc=page_sour.findAll('div',{'class':'_3Djpdu'})
    product_desc=desc[0].text
    description.append(product_desc)
    
    print("Product"+' ' + product_name)
    print("Price"+' ' + price)  
    print("description"+' ' + product_desc)
```

```python
df= pd.DataFrame({'Product':products,'Price':price,'Description':description})
df.to_csv("Mattress.csv")
bed=pd.read_csv("Mattress.csv")
bed.head()
```
