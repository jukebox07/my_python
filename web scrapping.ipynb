```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
my_url="https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG"
uClient=uReq(my_url)
page_html=uClient.read()
uClient.close()
```

```python
page_html
```

```python
page_soup=soup(page_html,"html.parser")
containers=page_soup.findAll("div",{"class":"_1YokD2 _3Mn1Gg col-12-12"})
print(len(containers))
```

```python
containers=page_soup.findAll("div",{"class":"_13oc-S"})
print(len(containers))
print(soup.prettify(containers[0]))
```

```python
container=containers[0]
image=container.findAll('div',{'class':'CXW8mj'})
image[0].img['alt']
```

```python
price=container.findAll('div',{'class':'_30jeq3 _1_WHN1'})
print(price[0].text)
```

```python
desc=container.findAll('div',{'class':'fMghEO'})
print(desc[0].text)
```

```python
ratings=container.findAll('span',{'class':'_2_R_DZ'})
print(ratings[0].text)
```

```python
products=[]
pricing=[]
rate=[]
description=[]

for container in containers:
    image = container.findAll('div',{'class':'CXW8mj'}) 
    product_name=image[0].img['alt']
    products.append(product_name)
    price_container=price=container.findAll('div',{'class':'_30jeq3 _1_WHN1'})
    price= price_container[0].text.strip()
    pricing.append(price)
    rating_container=container.findAll('span',{'class':'_2_R_DZ'})
    ratings=rating_container[0].text
    rate.append(ratings)
    desc=container.findAll('div',{'class':'fMghEO'})
    product_desc=desc[0].text
    description.append(product_desc)
    
    print("Product_Name"+' ' + product_name)
    print("Price"+' ' + price)  
    print("Ratings"+' ' + ratings)
    print("description"+' ' + product_desc)
        
    
    
```

```python
df= pd.DataFrame({'Product_Name':products,'Price':price,'Rating/Review':rate,'Description':description})
df.to_csv("Samsung_mobile.csv")
samsung=pd.read_csv("Samsung_mobile.csv")
samsung.head()
```
