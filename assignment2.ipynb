```python
Weather = pd.read_csv('C:\\Users\\User\\Desktop\\Datasets\\BicycleWeather.csv')
Weather.head()
```

```python
P = Weather['PRCP'].values
print("PRCP\n",P,"\n")

```

```python
inches = P/254 
inches.shape
```

```python
inches
val = pd.notna(inches)
inches[val]
```

```python
inche =inches[val]
non_zero = (inche > 0)
inche[non_zero]
```

```python
inche =inches[val]
zeros = (inche == 0)
inche[zeros]
```

```python
(inche[non_zero]) * 100 / 250
```

```python
print("Median precip on rainy days in BicycleWeather: ", np.median(inche[non_zero]))
print("Median precip on summer days in BicycleWeather: ", np.median(inche[zeros]))
print("Maximum precip on summer days in BicycleWeather: ", np.max(inche[zeros]))
print("Median precip on non-summer rainy days BicycleWeather: ", np.median(inche[non_zero & ~zeros]))
```
